```python
# imports
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/data copy.csv")
data.head(30)
```

```python
data['diagnosis'].value_counts()
```

```python
# list of variables and their data types
data.info()
```

```python
# central measurements
data.describe()
```

```python
# drop na's
data.drop(["Unnamed: 32", "id"], axis=1, inplace=True) # droping row which is axis=1 and inplace=True
```

```python
# one liner for for loop
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
data.head()
```

```python
# this is another way to convert categorical variables into binary encoding using numerical value 0 and 1
data['diagnosis'] = data['diagnosis'].astype("category", copy=False)
data['diagnosis'].value_counts().plot(kind="bar") # returns a value count of 0 and 1 for diagnosis column
```

```python
data.diagnosis
```

```python
# defining x predictors and y target variable
y = data["diagnosis"]
x = data.drop(["diagnosis"], axis=1)
```

```python
# normalizing data
from sklearn.preprocessing import StandardScaler

# create a scaler object
scaler = StandardScaler() # () means initalizing an oject

# fit the scaler to the data and then transform data
X_scaled = scaler.fit_transform(x)
X_scaled # all data points are as close to zero as possible
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.30, random_state=42)
```

```python
# train the model
from sklearn.linear_model import LogisticRegression

# create the lr model
lr = LogisticRegression()

# train the model on the training data
lr.fit(x_train, y_train)

# predict the target variable on test data
y_pred = lr.predict(x_test)
y_pred
```

```python
# evaluation of the model

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy: .2f}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
